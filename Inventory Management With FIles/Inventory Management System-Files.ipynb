## Product Details
1.Product ID  
2.Name of Product  
3.Price of Product  
4.Quantitiy Available  


```python
import time
#Reading the inventory and printing product detials
fd = open("inventory.txt","r")
products = fd.read().split("\n")
for product in products:
    product_details = product.split(",") 
    print(f"{product_details[0]} : {product_details[1]}")        
fd.close()

#Taking User Input
ui_username = input("Enter your Name : ")
ui_phone     = input("Enter Phone No : " )
ui_prod_id   = input("Enter Product ID : ")
ui_prod_qn   = input("Enter Quantitiy Needed : " )

updated_product_lst = []

#Going through each product detail
for product in products:
     
    product_details = product.split(",")    
    
    #checking whether the product exist in inventory
    if (product_details[0] == ui_prod_id):
        
        #checking whether stock available for requested quantitiy by user
        if(int(ui_prod_qn) <= int(product_details[3])):
            
            print("---------------------------------------------------")
            print("Product Name    :",product_details[1])
            print("Price per item  :",product_details[2])
            print("Quantity        :",ui_prod_qn)
            print("---------------------------------------------------")
            print(f"Total           : {int(product_details[2]) * int(ui_prod_qn)} /-")
            print("---------------------------------------------------")
            
            #updating Inventory List
            product_details[3] =  str(int(product_details[3]) - int(ui_prod_qn))
            
            #updating in sales.txt
            fd = open("Sales.txt","a")
            sales_detail = ui_username +","+ ui_phone +","+ product_details[1] +","+ ui_prod_id +","+ ui_prod_qn +","+ str(int(product_details[2]) * int(ui_prod_qn)) +","+ time.ctime()+"\n"
            fd.write(sales_detail)
            fd.close()
            
            
        # if we don't have requested quantity available   
        else:
            print("Sorry, We're not having enough quantity.")
            print("We're having only",product_details[3],'quantity.')
            print("Would you like to purchase it?")
            
            ch=input("Press Y/N : ")
            
            #if user want to purchase available quantity
            if(ch == "y" or ch == "Y"):
                print("---------------------------------------------------")
                print("Product Name    :",product_details[1])
                print("Price per item  :",product_details[2])
                print("Quantity        :",product_details[3])
                print("---------------------------------------------------")
                print(f"Total           : {int(product_details[2]) * int(product_details[3])} /-")
                print("---------------------------------------------------")
                
                #updating in sales.tx
                fd = open("Sales.txt","a")
                sales_detail = ui_username +","+ ui_phone +","+ product_details[1] +","+ ui_prod_id +","+ product_details[3] +","+ str(int(product_details[2]) * int(product_details[3])) +","+ time.ctime()+"\n"
                fd.write(sales_detail)
                fd.close()
                
                #updating Inventory List
                product_details[3] = "0"

            else:
                print("Sorry, We will update the stock shortly.")          
                
    #updated inventory List   
    updated_product_lst.append(product_details)
    
lst = []

#updating inventory string
for i in updated_product_lst:
    prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
    lst.append(prod)

lst[-1] = lst[-1][:-1]

#updating inventory file
fd = open("inventory.txt","w")
fd.writelines(lst)
fd.close()
```
